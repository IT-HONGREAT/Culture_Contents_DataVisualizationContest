## CNT_ 접두사 제거

* CNT_가 접두사로 있는 column명이 다수 존재해, 이를 수정하기 위해서
* 접두사를 제거하는 함수를 선언했습니다

```python
def drop_prefix(self, prefix):
    self.columns = self.columns.str.lstrip(prefix)
    return self

pd.core.frame.DataFrame.drop_prefix = drop_prefix
```

```python
clean_원본 = 제공데이터원본.drop_prefix('CNT_')
clean_원본.columns
```

```python
숫자형데이터 = clean_원본.iloc[:,6:]  #수치형변수들 전부
요일별데이터 = clean_원본.iloc[:,8:17]
사용처 = clean_원본.iloc[:,27:]
```

## 추정소득 Column 변경


* 문자형 등급방식으로 되어있던 기존의 추정소득 column을 숫자형으로 변경했습니다
( _오름차순으로 소득액이 높아집니다_ )


* [1]  -  A :	3천미만                
* [2]  -  B :	3 ~ 5천만원 미만           
* [3]  -  C :	5 ~ 7천만원 미만           
* [4]  -  D :	7천만 ~ 1억 미만

```python
clean_원본['추정소득']=clean_원본['추정소득'].apply(lambda x:x.replace("A","1"))
clean_원본['추정소득']=clean_원본['추정소득'].apply(lambda x:x.replace("B","2"))
clean_원본['추정소득']=clean_원본['추정소득'].apply(lambda x:x.replace("C","3"))
clean_원본['추정소득']=clean_원본['추정소득'].apply(lambda x:x.replace("D","4"))
clean_원본['추정소득']=clean_원본['추정소득'].apply(lambda x:x.replace("E","5"))

print("전체데이터 중 예시5개")
clean_원본.head()
```

```python
# 상관계수 함수 생성
def make_heatmap(inputdata):
    
    mask = np.zeros_like(inputdata.corr())
    mask[np.triu_indices_from(mask)] = True
    plt.rcParams["figure.figsize"] = (100,50) # 그림 크기 조정
    plt.rc('font', size=10)        # 기본 폰트 크기
    plt.rc('axes', labelsize=10)   # x,y축 label 폰트 크기
    plt.rc('xtick', labelsize=10)  # x축 눈금 폰트 크기 
    plt.rc('ytick', labelsize=10)  # y축 눈금 폰트 크기
    plt.rc('legend', fontsize=10)  # 범례 폰트 크기
    plt.rc('figure', titlesize=10) # figure title 폰트 크기
    
    return sb.heatmap(data = inputdata.corr(), mask=mask, annot=True, 
                      fmt = '.3f', linewidths=0, cmap='Greens')

```

```python
사용처상관지수 = MinMax_사용처.corr(method="pearson")
사용처상관지수

사용처상관지수통계지표 = 사용처상관지수.describe()
사용처상관지수통계지표
# 사용처상관지수통계지표.iloc[6,:]
```

```python
#각 직업별의 사용건이 높은 사용처 순으로 정렬하기(내림차순)

A =  직업_사용처.sort_values(by="A.전문직",ascending=False).head(10)
B =  직업_사용처.sort_values(by="B.회사원(대기업)",ascending=False).head(10)
C =  직업_사용처.sort_values(by="C.회사원(일반)",ascending=False).head(10)
D =  직업_사용처.sort_values(by="D.공무원",ascending=False).head(10)
E =  직업_사용처.sort_values(by="E.교육인",ascending=False).head(10)
F =  직업_사용처.sort_values(by="F.자영업자",ascending=False).head(10)
J =  직업_사용처.sort_values(by="J.기타",ascending=False).head(10)
```

## 직업별로 카드사용 사용처가 많은 순위

* 대부분의 직업분야에서 온라인쇼핑과 마트할인점등이 상위를 차지했습니다.

* 단, 자영업자와 기타직군은 '마트할인점'에서 '온라인쇼핑'보다 더 높은 사용건수를 보였습니다.

```python
직업들 = [C,J,F,B,D,E,A]   #데이터에서 많이 집계되었던 직업순
직업명 = ["회사원(일반)","기타","자영업자","회사원(대기업)","공무원","교육인","전문직"]
사용처순서dict = {}

for i,j in enumerate(직업들):
    직업별리스트 = list(j.index)    
    사용처순서dict[i] = 직업별리스트

직업별_사용처_top10= pd.DataFrame(사용처순서dict)
직업별_사용처_top10.columns=직업명
직업별_사용처_top10.index=["{}위".format(i) for i in range(1,11)]
직업별_사용처_top10
```

# 1인 가구 / 다 가구

* 1인가구와 다가구를 구분했습니다

* 원활한 변수지정을 위해 1인가구=일인가구 로 설정했습니다

* '일인가구여부'는 O,X로 구분합니다

```python
#일인가구 수O
clean_원본['일인가구여부']
```

```python
일인가구O = clean_원본['일인가구여부'] == 'O'
일인가구X = clean_원본['일인가구여부'] == 'X'
# 조건를 충족하는 데이터를 필터링하여 새로운 변수에 저장합니다.
일인가구 = clean_원본[일인가구O]
다가구 = clean_원본[일인가구X]
```

## 직업별 1인 가구

* 각 직업군에 1인 가구가 얼마나 있는지 확인했습니다.

```python
직업별_일인가구여부 = clean_원본.groupby(일인가구O)['직업'].value_counts()
직업별_일인가구 = 직업별_일인가구여부.iloc[7:]
직업순 = ["회사원(일반)","기타","자영업자","회사원(대기업)","공무원","교육인","전문직"]
비율리스트 = []

for i,j in zip(직업비율,직업별_일인가구):

    백분율 = round((j/i)*100,2)
    비율리스트.append(백분율)

for i,j in zip(직업명, 비율리스트):

    print("각 직업별 1인 가구의 비율 : ",i,"=>",j,'%')
    
직업별_일인가구_df = pd.DataFrame([비율리스트],columns=직업명)


직업별_일인가구_df
```

## 직업별 1인 가구의 비율

* 회사원 과 기타직 군에서 비교적 많은 비율로 1인 가구가 있었고
* 전문직, 교육인, 공무원은 매우 적은 수가 1인 가구 형태로 있었습니다.


```python
colors = sb.color_palette('husl',10)
f = plt.figure(figsize=(7,7))
plt.bar(직업순,비율리스트,color=colors)
plt.title('직업별_1인 가구')
plt.xlabel('직업')
plt.ylabel('각 직업별 1인 가구의 비율(%)')
plt.ylim(0,60)
plt.show()
```

## 연령성별 1인 가구

* 각 연령성별 층에 1인 가구가 얼마나 있는지 확인했습니다.

```python
연령성별비비율
연성순 = ["여_4050","남_30이하","남_4050","남_60이상","여_60이상","여_30이하"]
비율리스트 = []

연령성별_일인가구여부 = clean_원본.groupby(일인가구O)['연령성별'].value_counts()
연령성별_일인가구 = 연령성별_일인가구여부.iloc[6:]

for i,j in zip(연령성별비비율,연령성별_일인가구):

    
    백분율 = round((j/i)*100,2)
    비율리스트.append(백분율)

for i,j in zip(연성순, 비율리스트):
    print("각 연령성별당 1인 가구의 비율 : ",i,"=>",j,'%')

연령성별_일인가구_df = pd.DataFrame([비율리스트],columns=연성순)

연령성별_일인가구_df

```

## 연령성별 1인 가구의 비율

* 40~50대의 남성이 1위로 1인가구 비율이 제일 높았지만
* 모든 연령성별층이 직업별 비율에 비해 상대적으로 비슷한 비율을 보였습니다



```python
colors = sb.color_palette("husl", 5)
f = plt.figure(figsize=(7,7))
plt.bar(연성순,비율리스트,color=colors)
plt.title('연령성별_1인 가구')
plt.xlabel('연령성별')
plt.ylabel('각 연령성별당 1인 가구의 비율(%)')
plt.ylim(0,50)
plt.show()
```

## 소득별 1인 가구

* 각 소득별 층에 1인 가구가 얼마나 있는지 확인했습니다.

```python
소득별 = clean_원본.groupby(일인가구O)['추정소득'].value_counts()
추정소득순 = [2,1,3,4,5]
소득별.iloc[5:]

비율리스트 = []
for i,j in zip(소득별.iloc[:5],소득별.iloc[5:]):

    
    백분율 = round((j/(i+j))*100,2)
    비율리스트.append(백분율)

for i,j in zip(추정소득순, 비율리스트):
    print("각 소득순 별 1인 가구의 비율 : ",i,"=>",j,'%')

소득별_일인가구_df = pd.DataFrame([비율리스트],columns=추정소득순)

소득별_일인가구_df
```

## 상관관계 탐색

* 추정소득_온라인쇼핑에 대해
* 상관관계를 찾기 위한 탐색을 시행했으나, 상호 연관된 데이터로 보여지지 않았습니다

```python
추정소득col = clean_원본['추정소득']
온라인쇼핑col = clean_원본['온라인쇼핑']


df_concat = pd.concat([추정소득col,온라인쇼핑col],axis=1)
scaler = MinMaxScaler()
scaler.fit(df_concat)
scaled_df_concat = scaler.transform(df_concat)

f = plt.figure()
추정소득_온라인쇼핑 = pd.DataFrame(data=scaled_df_concat,columns = df_concat.columns)

추정소득_온라인쇼핑.plot(kind='scatter',x='추정소득',y='온라인쇼핑')

```

```python
from sklearn.preprocessing import RobustScaler

추정소득col = clean_원본['추정소득']
온라인쇼핑col = clean_원본['온라인쇼핑']


df_concat = pd.concat([추정소득col,온라인쇼핑col],axis=1)
scaler = RobustScaler()
scaler.fit(df_concat)
scaled_df_concat = scaler.transform(df_concat)

f = plt.figure()
추정소득_온라인쇼핑 = pd.DataFrame(data=scaled_df_concat,columns = df_concat.columns)

추정소득_온라인쇼핑.plot(kind='scatter',x='추정소득',y='온라인쇼핑')

```
